```python
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/world_population dataset (Mayuri).csv')
df
```

```python
del df['CCA3']  # Deleting the CCA3 column
```

```python
df=df.rename(columns={'Area (km²)':'Area'})
df
df=df.rename(columns={'Density (per km²)':'Density'})
df
```

```python
# 1. Which country/territory experienced the highest population growth between 2010 and 2022?

population_columns = ['2010 Population', '2022 Population']
df['Population Growth'] = df['2022 Population'] - df['2010 Population']

max_growth_country = df.loc[df['Population Growth'].idxmax(), 'Country/Territory']
print("The country/territory with the highest population growth between 2010 and 2022 is:" ,max_growth_country)
df_sorted = df.sort_values(by='Population Growth', ascending=False)

# Plotting
plt.figure(figsize=(12, 8))
bar_plot = plt.bar(df_sorted['Country/Territory'], df_sorted['Population Growth'], color='skyblue')
plt.xlabel('Country/Territory')
plt.ylabel('Population Growth (2010-2022)')
plt.title('Population Growth Between 2010 and 2022 for Each Country/Territory')

label_indices = range(0, len(df_sorted['Country/Territory']), 5)
plt.xticks(label_indices, df_sorted['Country/Territory'].iloc[label_indices], rotation=90, ha='right', fontsize=8)

plt.tight_layout()
plt.show()
```

```python
# 2.Identify the top three countries/territories with the largest land area in the dataset.

df_sorted_area = df.sort_values(by='Area (km²)', ascending=False)
top_three_area = df_sorted_area[['Country/Territory', 'Area (km²)']].head(3)

print("Top Three Countries/Territories with the Largest Land Area:")
top_three_area
```

```python
# 3. Identify any country/territory that experienced a decrease in population between 2015 and 2022.

df['Population Change (2015-2022)'] = df['2022 Population'] - df['2015 Population']

# decrease in population countries
decreasing_population = df[df['Population Change (2015-2022)'] < 0].head(10)

print("Countries/Territories with a Decrease in Population (2015-2022):")
print(decreasing_population[['Country/Territory', 'Population Change (2015-2022)']])
```

```python
df['Population Change (2015-2022)'] = df['2022 Population'] - df['2015 Population']
decreasing_population = df[df['Population Change (2015-2022)'] < 0]

# Plotting
plt.figure(figsize=(12, 8))
bar_plot = plt.bar(decreasing_population['Country/Territory'], decreasing_population['Population Change (2015-2022)'], color='coral')
plt.xlabel('Country/Territory')
plt.ylabel('Population Change (2015-2022)')
plt.title('Population Change Between 2015 and 2022 for Countries/Territories with Decrease')
plt.xticks(rotation=90, ha='right', fontsize=8)  # Rotate and adjust font size for x-axis labels
plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

```python
# 4. Can you identify a country in the dataset that changed its capital city over the years?

rows_per_country = df.groupby('Country/Territory').size()
changing_capital_countries = rows_per_country[rows_per_country > 1]
print("Countries with multiple rows indicating potential capital changes:")
print(changing_capital_countries)
```

```python
# 5. Explore the relationship between the growth rate and the current population for various countries.
scatter_data = df[['Country/Territory', '2022 Population', 'Growth Rate']]

# Plotting the scatter plot
plt.figure(figsize=(12, 8))
plt.scatter(scatter_data['2022 Population'], scatter_data['Growth Rate'], color='skyblue', alpha=0.5, label='Countries')
plt.title('Relationship Between Growth Rate and Current Population for Various Countries')
plt.xlabel('2022 Population')
plt.ylabel('Growth Rate')
plt.grid(True)
plt.legend()
plt.show()

```

```python
scatter_data = df[['Country/Territory', '2022 Population', 'Growth Rate']]

# Plotting the 2D histogram (histplot)
plt.figure(figsize=(12, 8))
sns.histplot(data=scatter_data, x='2022 Population', y='Growth Rate', bins=20, cmap='Blues', cbar=True)
plt.title('2D Histogram: Relationship Between Growth Rate and Current Population')
plt.xlabel('2022 Population')
plt.ylabel('Growth Rate')
plt.show()

```

```python
# 6. Describe Top 5 capital with highest rank ?
sort=df.sort_values(by='Rank',ascending=False)
sort.head(5)
```

```python
top_capitals_data = df.nlargest(5, 'Rank')[['Country/Territory', 'Capital', 'Rank']]

# Plotting the pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_capitals_data['Rank'], labels=top_capitals_data['Capital'], autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)
plt.title('Distribution of Rank for Top 5 Capitals')
plt.show()
```
